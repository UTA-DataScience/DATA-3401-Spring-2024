### **Question 1:**

## Creature and Pokemon Hierarchy

### Objective:
 You will create a superclass named `Creature` and a subclass named `Pokemon` which extends the functionality of `Creature`.

#### Tasks:

1. **Creature Class**:
   - Create a superclass named `Creature` with the following attributes:
       - `name` (string)
       - `health` (integer)
   - Implement an `__init__` method that initializes a `Creature` with `name` and `health`.
   - Include the following methods in the `Creature` class:
       - `display_info()`: Prints the creature's `name` and `health`.
       - `take_damage(damage)`: Reduces the `health` by the `damage` amount and prints the creature's `name` and updated `health`. If `health` falls at or below 0, print that the creature has fainted.
       - `heal(amount)`: Increases the `health` by the `amount` specified and prints the creature's `name` and updated `health`.

2. **Pokemon Class**:
   - Create a subclass named `Pokemon` that inherits from `Creature`. Add the following additional attributes:
       - `poke_type` (string)
       - `level` (integer)
       - `power` (integer)
       - `speed` (integer)
   - Override the `__init__` method to include initialization of the new attributes and ensure it still initializes the inherited attributes.
   - Override the `display_info()` method to include all the attributes from the `Pokemon` class.
   - Implement an `attack(other_creature)` method that allows a `Pokemon` to attack another `Creature` or `Pokemon`. The method should calculate damage based on the attacker's `level` and `power` and then call `other_creature.take_damage(damage)` with the calculated damage.
   - Implement a `speed_check(other_creature)` method that compares the `speed` of two creatures and prints which one would move first based on their `speed`.

3. **Testing Your Code**:
   - Create two `Pokemon` instances with different attributes and demonstrate each method, including inheritance and method overriding by interacting with both `Creature` and `Pokemon` instances.

For Example:
- pikachu = Pokemon("Pikachu", 100, "Electric", 5, 50, 60)

Good luck, and may the best Pokemon win!

### **Question 2:**

## - The Vehicle Hierarchy


### Tasks:

1. **Vehicle Class (Superclass):**
   - Define a superclass named `Vehicle` with the following attributes:
       - `make` (string): The manufacturer of the vehicle.
       - `model` (string): The model of the vehicle.
       - `year` (integer): The manufacturing year of the vehicle.
       - `fuel_type` (string): The type of fuel the vehicle uses (e.g., gasoline, diesel, electric).
   - Implement an `__init__` method that initializes a `Vehicle` with the attributes listed above.
   - Include the following methods in the `Vehicle` class:
       - `display_info()`: Prints the vehicle's details, including make, model, year, and fuel type.
       - `age()`: Calculates and returns the age of the vehicle based on the current year.

2. **Car Class (Subclass):**
   - Create a subclass named `Car` that inherits from `Vehicle`. Add the following additional attributes:
       - `passenger_capacity` (integer): The number of passengers the car can carry.
       - `trunk_size` (float): The size of the car's trunk in cubic feet.
   - Override the `__init__` method to include initialization of the new attributes while ensuring the inherited attributes are also initialized.
   - Override the `display_info()` method to include the passenger capacity and trunk size.

3. **Truck Class (Subclass):**
   - Create a subclass named `Truck` that inherits from `Vehicle`. Add the following additional attribute:
       - `cargo_capacity` (float): The cargo capacity of the truck in pounds.
       - `four_wheel_drive` (boolean): Whether the truck has four-wheel drive capability.
   - Override the `__init__` method to include initialization of the new attributes while ensuring the inherited attributes are also initialized.
   - Override the `display_info()` method to include cargo capacity and whether it has four-wheel drive.

4. **Testing Your Code:**
   - Create instances of `Car` and `Truck` with relevant attributes.
   - Demonstrate the use of each method, including inherited and overridden methods, by interacting with both `Car` and `Truck` instances.
