```python
class pizza:
  def __init__(self, name):
    self.name=name
    self.topping=[]

  def add_topping(self, topping): # method
    self.topping.append(topping)

  def describe(self):
    return f"{self.name} pizza with toppings: '{','.join(self.topping)}"
```

```python
# Object 1
margherita=pizza("margherita")
margherita.add_topping("Tomato")
margherita.add_topping("Cheese")
margherita.describe()
```

```python
## Example: Car

class Car:
  def __init__(self, color, brand, speed):
    self.color=color
    self.brand=brand
    self.speed=speed
    self.fuel_level=100

  def drive(self):
    self.fuel_level -=10
    print("Car is driving")

  def stop(self):
    print("Car stopped")

  def refuel(self):
    self.fuel_level=100
    print("Car refueled")
```

```python
my_car=Car("red", "BMW", 110)
my_car.drive()
my_car.stop()
my_car.refuel()
```

```python
#### Example: Class

class Book:
  def __init__(self, title, author, number_pages, genre):
    self.title=title
    self.author=author
    self.number_pages=number_pages
    self.genre=genre
    self.current_page=1

  def open(self):
    print(f"{self.title} open to page {self.current_page}")

  def close(self):
    print(f"{self.title} closed")

  def read_pages(self):
    if self.current_page<=self.number_pages:
      print(f"reading page {self.current_page}")
      self.current_page+=1

    else:
      print("End of the book")
```

```python
my_book=Book("Gatsby", "Scott", 180, "Fiction")
```

### Example: Smartphone

```python
class Smartphone:
  def __init__(self,brand, model, color, battery_level, storage_capacity):
    self.brand=brand
    self.model=model
    self.color=color
    self.battery_level=battery_level
    self.storage_capacity=storage_capacity

  def make_call(self,number):
    if self.battery_level>0:
      self.battery_level-=1
      print(f"Calling{number}...")
    else:
      print("Phone is dead, please charge the phone")

  def send_message(self, number,message):
    if self.battery_level>0:
      self.battery_level-=1
      print(f"Sending messaafe to {number}: {message}")
    else:
      print("Phone is dead, please charge the phone")

  def take_phot(self):
    if self.battery_level>0:
      self.battery_level-=5
      print("Photo taken")

    else:
      print("Phone is dead, please charge the phone")

  def charge(self):
    self.battery_level=100
    print("Phone is fully charged")


```

```python
phone1=Smartphone("Apple", "Iphone14", "red",50, 256)
```

```python
phone1.take_phot()
```

```python
phone1.make_call("12344556")
phone1.send_message("12344556", "Hi")
```

## **Inheritance**

```python
## Create super class (Parent Class)

class pizza:
  def __init__(self, name):
    self.name=name
    self.topping=[]

  def add_topping(self, topping): # method
    self.topping.append(topping)

  def describe(self):
    return f"{self.name} pizza with toppings: '{','.join(self.topping)}"
```

```python
# child class or Derived class

class MargheritaPizza(pizza):
  def __init__(self):
    super().__init__("Margaritha")
    self.add_topping("Tomato")
    self.add_topping("Cheese")
```

```python
class PepperoniPizza(pizza):
  def __init__(self):
    super().__init__("Pepperoni")
    self.add_topping("Pepperoni")
    self.add_topping("Olive")
    self.add_topping("Tomato sauce")
```
