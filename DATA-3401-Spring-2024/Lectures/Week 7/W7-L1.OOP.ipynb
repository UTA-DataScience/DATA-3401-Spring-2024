### **List Comprehensions**

```python
cities=["Dallas", "Arlington", "Denver", "Amherst"]

cap_cities=[]

for city in cities:
  cap_cities.append(city.upper())

print(cap_cities)
```

```python
capi_cities=[city.upper() for city in cities]
capi_cities
```

```python
### Another Example

squares=[]

for x in range(12):
  squares.append(x**2)

print(squares)
```

```python
squares2=[x**2 for x in range(12)]
squares2
```

```python
## add the condition

squares2=[x**2 for x in range(12) if x%2==0]
squares2
```

```python
####

squares3=[x**2 if x%2==0 else x+3 for x in range(20)]
squares3
```

```python
#### Create a class: Book

class Book:
  def __init__(self, title, author, pages, genre):
    self.title=title
    self.author=author
    self.pages=pages
    self.genre=genre

  def describe(self):
    return f" '{self.title}' by {self.author} has {self.pages} pages and belong to the {self.genre} genre."


```

```python
book1=Book("The gReat Gatsby", "Scott", 200, "Fiction")
```

```python
book2=Book("History of time", "Hawking", 300, "Science")
```
