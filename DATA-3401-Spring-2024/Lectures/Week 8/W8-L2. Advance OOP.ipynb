```python
import math

class ListMath:
    def __init__(self, numbers):
        self.numbers = numbers

    def display(self):
        print("Current List:", self.numbers)

    def max_value(self):
        return max(self.numbers)

    def min_value(self):
        return min(self.numbers)

    def average(self):
        return sum(self.numbers) / len(self.numbers)

    def standard_deviation(self):
        mean = self.average()
        variance = sum([(x - mean) ** 2 for x in self.numbers]) / len(self.numbers)
        return math.sqrt(variance)

    def square_root_transform(self):
        self.numbers = [math.sqrt(x) for x in self.numbers if x >= 0]
        print("After square root transformation:", self.numbers)

    def power_transform(self, power):
        self.numbers = [x ** power for x in self.numbers]
        print(f"After raising to the power of {power}:", self.numbers)

    def filter_negative(self):
        self.numbers = [x for x in self.numbers if x >= 0]
        print("After filtering negative numbers:", self.numbers)
```

```python
data=ListMath([1,4,-2,8,12,-34,56])
```

```python
data.display()
```

```python
data.max_value()
```

```python

print("the average of these numbers are:", data.average())
```

```python
data.standard_deviation()
```

```python
data.power_transform(2)
```

## **Example 2: A University System**

```python
from collections.abc import AsyncGenerator
#### Base Class: Person

class Person:
  def __init__(self, name, age):
    self.name=name
    self.age=age
  def display_info(self):
    print(f"Name: {self.name}, Age: {self.age}")

  def celebrate_birthday(self):
    self.age += 1
    print(f"Happy Birthday, {self.name}! you are now {self.age} years old")
```

```python
class Student(Person):
  def __init__(self,name,age,student_id):
    super().__init__(name,age)
    self.student_id=student_id
    self.courses=[]

  def enroll(self,course):
    if course not in self.courses:
      self.courses.append(course)
      print(f"{self.name} add in this course {course}")
    else:
      print(f"{self.name} alraedy enrolled in {course}")

  def display_courses(self):
    if self.courses:
      print(f"{self.name} courses")
      for course in self.courses:
        print(f" - {course}")
    else:
      print(f"{self.name} is not enrolled in these courses")
```

```python
class Instructor(Person):
  def __init__(self, name,age, emploee_id):
    super().__init__(name,age)
    self.emploee_id=emploee_id
    self.courses_taught=[]

  def assign_teaching(self, course):
    if course not in self.courses_taught:
      self.courses_taught.append(course)
      print(f"{self.name} is now teaching {course}")

  def display(self):
    print(f"{self.name} course taught:")
    for course in self.courses_taught:
      print(f"- {course}")

```

```python
alice=Student("Alice", 22, "344567")
Tom=Instructor("Bob", 46, "98876655")
```

```python
alice.celebrate_birthday()
```

```python
Tom.assign_teaching("Math")
```

```python
import pandas as pd
```

```python
class DatasetAnalyzer:
  def __init__(self,filepath):
    self.df=pd.read_csv(filepath)

  def missing_values(self):
    return self.df.isnull().sum()

  def duplicates(self):
    return self.df.duplicated().sum()

  def dimension(self):
    return self.df.shape

  def head(self,rows=5):
    return self.df.head(rows)
```

```python
analyzer=DatasetAnalyzer("data.csv")
```
