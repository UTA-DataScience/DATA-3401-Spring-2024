```python
class Pizza:
  def __init__(self, name):
    self.name=name
    self.toppings=[]

  def add_topping(self, topping):
    self.toppings.append(topping)

  def bake(self):
    print(f"Baking {self.name} pizza with toppings : {','.join(self.toppings)}..")

## Subclass
class MargheritaPizza(Pizza):
  def __init__(self):
    super().__init__("Margherita")
    self.add_topping("Tomato")
    self.add_topping("Cheese")
```

```python
### Create an instance
Margherita=MargheritaPizza()
Margherita.bake()
```

```python
class Vehicle:
  def __init__(self, make, year,mileage):
    self.make=make
    self.year=year
    self.mileage=mileage
  def display_info(self):
    print(f"Vehivle Infor: {self.year} {self.make}, Mileage: {self.mileage} miles")

  def start_engine(self):
    print("Engine Started")
  def set_mileage(self,mileage):
    if mileage>=self.mileage:
      self.mileage=mileage
      print(f"New Mileage set to {self.mileage} miles")
    else:
      print("Error: ?????")
  def get_mileage(self):
    return self.mileage

  def drive(self, distance):
    if distance>0:
      self.mileage+=distance
      print(f"Driving {distance} miles. the New Mileage is {self.mileage} miles")
    else:
      print("Distance should be greater than Zero")
```

```python
class Car(Vehicle):
  def __init__(self, make, year, mileage, doors, fuel_type):
    super().__init__(make,year,mileage)
    self.doors=doors
    self.fuel_type=fuel_type

  def display_info(self):
    super().display_info()
    print(f"Doors: {self.doors}, Fuel type: {self.fuel_type}")
  def start_engine(self):
    if self.fuel_type.lower()=="electric":
      print("Car Engine start silently")
    else:
      print("Car Engine start with a roar")

  def lock_doors(self):
    print("Door Locked")
  def unlock_doors(self,code):
    if code=="12345":
      print("Door Unlocked")
    else:
      print("Wrong Code")
```

```python
car=Car("Toyota", 2020, 25000,4, "Petrol")
car.display_info()
```

```python
car.start_engine()
```

```python
car.drive(100)
```

```python
car.unlock_doors("12345")
```

```python
class ListMath:
  def __init__(self,numbers):
    self.numbers=numbers
  def display(self):
    print("Current Lists", self.numbers)
  def max_value(self):
    return max(self.numbers)
  def min_value(self):
    return min(self.numbers)
  def average(self):
    return sum(self.numbers)/len(self.numbers)
  def standard_deviation(self):
    mean=self.average()
    variance=sum([(x-mean)**2 for x in self.numbers])/len(self.numbers)
    return math.sqrt(variance)

  def square_root_transform(self):
    self.numbers=[math.sqrt(x) for x in self.numbers if x>=0]
    print("After square root transfromation:", self.numbers)

  def power_transform(self, power):
    self.numbers=[x ** power for x in self.numbers]
    print("Hi")
  def filter_negative(self):
    self.numbers=[x for x in self.numbers if x>=0]
    print("After filtering the negative numbers",self.numbers )
```
