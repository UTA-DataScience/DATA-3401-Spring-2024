```python
### chage the size of all the graphs
plt.rcParams["figure.figsize"]=6,6

x=np.linspace(0,10,100)
y=np.sin(x)

fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

for ax in axes:
  ax.plot(x,y)
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_title("Graphs")

plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(8,6))

fig.tight_layout()

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title("Original Graph")

axes[1].plot(x,x**2,x,x**3)
axes[1].set_title("Custom range Graph")
axes[1].set_xlim(2,4)
axes[1].set_ylim(0,60)

axes[2].plot(x,x**2,label='$y=x^2$')
axes[2].plot(x,x**3,label='$y=x^3$')
axes[2].set_title("Custom Ticks and Labels")
axes[2].set_xlim(2,4)
axes[2].set_ylim(0,60)
axes[2].set_xticks([2,2.5,3,3.5,4])
axes[2].set_yticks([0,20,40,60])

plt.show()

```

### Using Loop to Create the graphs

```python
data=sns.load_dataset('mpg')

for column in data.select_dtypes(include=['float64','int64']).columns:
  plt.figure()
  sns.histplot(data[column], kde=True)
  plt.title(f'Histogram of {column}')
  plt.show()
```

```python


for column in data.columns:
  plt.figure()
  if data[column].dtype=='object':
    sns.countplot(y=column, data=data)
    plt.title(f'Count plot chart for {column}')

  else:
    sns.histplot(data[column], kde=True)
    plt.title(f'Histogram plot chart for {column}')
  plt.show()
```
