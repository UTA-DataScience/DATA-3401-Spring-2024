```python
a={"x_position":0, "y_position":25, "speed": "fast"}
```

```python
if a["speed"]=="slow":
  x_incremnt=1
elif a["speed"]=="medium":
  x_incremnt=2
else:
  x_incremnt=3
```

```python
## The new position is the old position plus the increment
a["x_position"]=a["x_position"]+x_incremnt
print(f"New Position: {a['x_position']}")
```

```python
#### Looping through the keys and values

user={'username': 'leo123',
      'first':'Leo',
      'last': 'Messi'}
user
```

```python
for key,value in user.items():
  print(f"\nkey: {key}")
  print(f"\nvalue: {value}")
```

```python
### Create another dictionary

my_languages={"jen": "Python",
              "Sarah": "C",
              "Jeff": "R",
              "Al": "Java",
              "Bob": "SQL"}
```

```python
for name in my_languages.keys():
  print(name.upper())
```

```python
for lan in my_languages.values():
  print(lan.upper())
```
