```python
### Lets find sum of positive numbers

def sum_of_positives(numbers):
  total=0
  for number in numbers:
    if number>0:
      total +=number
  return total
```

```python
numbers=[1,4,-5,12,34,31,6]
```

```python
sum_of_positives(numbers)
```

```python
#### Create a function that find the positive numbers

def filter_positive_numbers(numbers):
  positive_numbers=[]
  for number in numbers:
    if number>0:
      positive_numbers.append(number)
  return positive_numbers
```

```python
numbers=[-10,-20,50,34,12]
```

```python
filter_positive_numbers(numbers)
```

```python
##### Replace even numbers with their square

def square_even_numbers(numbers):
  result=[]
  for number in numbers:
    if number%2==0:
      result.append(number**2)
    else:
      result.append(number)
  return result

numbers=[1,3,5,11,2,8,6,88]
```

```python
square_even_numbers(numbers)
```

```python
### Create a function that count negative numbers

def count_negatives(numbers):
  count=0
  for number in numbers:
    if number <0:
      count+=1
  return count
```

```python
numbers=[1,3,-5,-11,2,8,6,-88]
count_negatives(numbers)
```

```python
 #### Create a function to find the maximum value
def find_max_v(values):
  max_value=values[0]
  for value in values:
    if value>max_value:
      max_value=value
  return max_value
```

```python
values=(1,4,6,7,8)
```

```python
find_max_v(values)
```
