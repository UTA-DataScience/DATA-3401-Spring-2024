## **While:**

```python
current_number=1

while current_number<=5:
  print(current_number)
  current_number=current_number+1 # current_number+=1
```

```python
## Another Example

n=10

while n>0:
  print("The value of n is:", n)
  n-=1 # n=n-1
```

```python
## Another example

number=1

while number<50:
  number*=2 # numer=number*2
  print(number)
```

```python
### Lets choose when user can quit

message=" "

while message != 'quit':
  message=input()
  print(message)
```

```python
### Another example


count=0

while count <5:
  print("Count:", count)
  count+=1
```

#### **Break and Continue**

```python
n=0

while True:
  print(n)
  n+=1
  if n>=5:
    break
```

```python
## Another example

while True:
  city=input()
  if city =="Dallas":
    break
  else:
    print(f"I love to go to {city.title()}")
```

```python
### Continue

# Lets skip the even numbers and print only odd numbers between 1 and 10

x=0
while x<10:
  x+=1
  if x%2==0:
    continue
  print(x)
```

```python
### for loop and conditional statement

numbers=[1,2,3,4,5]

for i in numbers:
  if i%2==0:
    print(f"{i} is even")
  else:
    print(f"{i} is odd")
```

```python
### Nested for loop

for j in range(3):
  for i in range(3):
    print(j,i)
```
