### **Nested If statement**

```python
## Lets check for age groups

## Divide the age to different groups

age=40

if age<18:
  if age <13:
    print("Child")
  else:
    print("Teenager")
else:
  if age<65:
    print("Adult")
  else:
    print("Senior")
```

```python
### Checking for grades

score=85
if score >=90:
  print("Grade A")
else:
  if score>=80:
    print("Grade B")
  else:
    if score>=70:
      print("Grade C")
    else:
      if score>=60:
        print("Grade D")
      else:
        print("F")
```

```python
### Checking type of triangle based on side lengths:

side1=3
side2=4
side3=5

if side1+side2>side3:
  if side1==side2==side3:
    print("Equilateral Triangle")
  elif side1==side2 or side2==side3 or side1==side3:
    print("Isosceles Triangle")
  else:
    print("Scalene Triangle")
else:
  print("Not a valid triangle")

```

#### Logical Operators: **and**

```python
x=5
y=10

if x>0 and y>0:
  print("Both x and y are positive")
else:
  print("atlease one of them are negative")
```

#### Logical Operators: **or**

```python
age=25

if age<18 or age>=65:
  print("A")
else:
  print("B")
```

#### Logical Operators: **not**

```python
is_raining=True

if not is_raining:
  print("It is not raining")
else:
  print("it is raining")
```

```python
### Combing logical operators

age=30
income=50000

if (age>=30 and age<=65) and income>30000:
  print("A")
else:
  print("B")
```
