```python
### Round the return value to 2

def cylinder_volume (radius, height):
  volume=3.14*radius**2*height

  rounded_volume=round(volume,2)

  return rounded_volume
```

```python
cylinder_volume(10.2,8)
```

```python
def add (a,b):
  return a+b

def multiply (a,b):
  return a*b

def add_and_multiply(x,y,z):
  addition_result=add(x,y)
  final_result=multiply(addition_result,z)
  return final_result
```

```python
add_and_multiply(3,5,7)
```

### **Functions with positional and keyword Arguments**

```python
def describe_pet(animal_type, pet_name, age=None):
  print(f"I have a {animal_type} named {pet_name}")
  if age:
    print(f"it is {age} yeras old")
  else:
    print("the age is unkown")
```

```python
describe_pet("Hamster", "harry")
```

```python
describe_pet("Hamster", "harry", 6)
```

### **Arbitrary Arguments:**

```python
## Example 1

def example_function(*args):
  for arg in args:
    print(arg)
```

```python
### Build a sandwich

def build_sandwich(size,*fillings):
  print("Making a " +size + " sandwich with the following fillings")

  for filling in fillings:
    print("-", filling)
```

```python
build_sandwich("Large", "pesto", "mayo","tomato", "lettuce", "beef","cheese")
```
