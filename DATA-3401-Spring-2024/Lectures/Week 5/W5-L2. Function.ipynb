## **Positiona aribitrary Arguments**

```python
def make_pizza(*toppings):
  print(toppings)
```

```python
make_pizza("pepperoni", "cheese", "tomato", "mayo")
```

```python
def make_pizza(*toppings):
  for topping in toppings:
    print("-" + topping)
```

```python
make_pizza("pepperoni", "cheese", "tomato", "mayo")
```

```python
### Create a vacation package

def book_vacation(destination,days,*activities,hotel="Standard",transport="Economy"):
  print(f"Booking a {days}-day trip to {destination}")
  print(f"Hotel: {hotel}")
  print(f"Transport {transport}")
  print("Planned activities are include:")
  for activity in activities:
    print("-"+activity)
```

```python
book_vacation("Hawaii",14, "surfing", "hiking","Swiming",hotel="Luxury",transport="First Class")
```

```python
book_vacation("surfing", "hiking","Swiming", "Hawaii",14,hotel="Luxury",transport="First Class")
```

```python
### Organizing a seminar

def organize_seminar(date,*speakers, topic="general", duration=2, venue="Main Hall"):
  print(f"Organizing a seminar on {topic} on {date} for {duration} hours")
  print(f"Venue: {venue}")
  print("Speakers include:")
  for speaker in speakers:
    print("-" +speaker)
```

```python
organize_seminar("10th October", "Dr.Smith", "Dr. Leckie", "Dr. Mo",
                 topic="Data Science", duration=6, venue="Science HAll")
```

# ** Aribitrary keyword Arguments

```python
## Create a function for personal detail

def personal_details(**details):
  for key,value in details.items():
    print(f"{key.capitalize()}: {value}")
```

```python
personal_details(name="John", age=32, city="New Yord")
```

```python
### Calculate the area and Perimeter of Rectangle

def rectangle(l,w):
  def area():
    return l*w

  def perimeter():
    return 2*(l+w)

  return (area(),perimeter() )
```

```python
a,p=rectangle(3,5)

print(f"Area :{a}, perimeter: {p}")
```

```python
### generate a function ro raise the number to specific power

def power(n):
  def compute(x):
    return x**n
  return compute
```

```python
square=power(2)
cube=power(3)

print((square(4)))
print(cube(3))
```

```python
multiply(3,2)
```

```python
### Use Lambda with filter()

numbers=[1,3,5,7,8,12,34,56,34,2,1]

even_numbers=list(filter(lambda x: x%2==0, numbers))

print(even_numbers)

```
