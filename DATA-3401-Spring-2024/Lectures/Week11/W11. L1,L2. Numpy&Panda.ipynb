### **Linear Algebra**


```python
## MM

a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
matrix_product=np.dot(a,b)
print(matrix_product)
```

```python
### Linear system of equations

d=np.array([[1,2],[3,4]])
o=np.array([5,6])

solutions=np.linalg.solve(d,o)
print(solutions)
```

#### **Data Cleaning**

```python
data={
    "Name":["Alice", "Leo","Charlie", "David", np.nan],
    "Age":[26,34,np.nan,14,56],
    "City":["New York","Arlington","Dallas","Fort Worth", "Orlando"],
    "Salary":[45000,56000,32000,90000,49000]
}

df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
## How to handle the missign values

df1=df.dropna()
df1
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
#### fill your missing values with specific values
df_filled=df.fillna({"Name":"Unknown", "Age":0})
df_filled
```

```python
### Imputation

df["Age"]=df["Age"].fillna(df["Age"].median())
df
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
### Transfrom the data types

df["Age"]=df["Age"].astype(int)
df.info()
```
